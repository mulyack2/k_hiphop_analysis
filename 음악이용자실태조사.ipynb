```python
genres = [
    "Ballad",
    "Dance",
    "OST",
    "R&B/Soul",
    "Rap/Hip-Hop",
    "Classic",
    "Indie",
    "Rock/Metal",
    "Trot",
    "Jazz",
    "EDM",
    "Other",
]
```

```python
total_music_df = music_df[music_df["Category"] == "Total"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["font.family"] = "AppleGothic"
```

```python
# 음악시장 청취자 크기

total_df = total_music_df.set_index("date").loc[:, genres].sum(axis=1).rename("total").reset_index()

fig, ax = plt.subplots(figsize=(14, 6))
sns.barplot(x="date", y="total", data=total_df, palette="tab10", hue="date", legend=True, ax=ax)
for container in ax.containers:
    ax.bar_label(container, fmt="Value:\n%d", label_type="center")
plt.show()
```

```python
# 시대별 / 장르별
pivot_df = total_music_df.drop(columns=["Category"]).set_index("date").T
pivot_df.reset_index(names="Category", inplace=True)

df_melted = pivot_df.melt(id_vars="Category", var_name="date", value_name="total")

plt.figure(figsize=(14, 6))
sns.barplot(x="date", y="total", hue="Category", data=df_melted, palette="tab10")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 장르별 / 시대별
fig, axs = plt.subplots(1, 2, figsize=(14, 6)) 
sns.barplot(x="Category", y="total", hue="date", data=df_melted[df_melted['Category'].isin(genres[:6])], palette="tab10",ax=axs[0])
sns.barplot(x="Category", y="total", hue="date", data=df_melted[df_melted['Category'].isin(genres[6:])], palette="tab10",ax=axs[1])
plt.tight_layout()
plt.show()
```
